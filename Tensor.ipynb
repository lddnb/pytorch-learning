# Tensor

```python
import torch
import numpy as np
```

## 张量初始化

```python
# type 1 : Directly from data
data = [[1, 2],[3, 4]]
x_data = torch.tensor(data)

# type 2 : From a NumPy array
np_array = np.array(data)
x_np = torch.from_numpy(np_array)

# type 3 : From another tensor
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")

print(f"Origin Tensor: \n {x_data} \n")
print(f"Shape : {x_data.shape}")
```

## 随机或常量初始化

```python
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

## 张量属性

```python
tensor = torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## 张量运算

```python
tensor = torch.ones(2, 3)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
# tensor = tensor.to(device)
print(f"Device tensor is stored on: {tensor.device}")
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
tensor[:,1] = 0
print(tensor)
```

### 张量拼接

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1) # 对于row和col的第二维拼接
t2 = torch.cat([tensor, tensor], dim=0) # 对于row和col的第一维拼接
print(t1)
print(t2)
```

### 算数运算

```python
# 三种等价的矩阵乘法
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3)
print(y3)

# 三种等价的逐元素相乘
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
print(z3)

# 逐元素求和
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))

# 原地操作
print(f"{tensor} \n")
tensor.add_(5)
print(tensor)
tensor.copy_(3)
print(tensor)
# 原地操作节省了一些内存，但在计算导数时可能存在问题，因为历史会立即丢失。因此，不建议使用
```

## 与 NumPy 的转换

```python
# Tensor to NumPy array
t = torch.ones(3)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")

t.add_(1)
print(f"t: {t}")
print(f"n: {n}")

np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
# 这里两者的变化会相互影响
```

```python
# NumPy array to Tensor
n = np.ones(5)
t = torch.from_numpy(n)
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")

t.add_(1)
print(f"t: {t}")
print(f"n: {n}")

# 这里两者的变化同样会相互影响
```
